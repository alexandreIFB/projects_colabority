```python
import re

filename = "/content/resultado_blast"

fp = open(filename)
#transformo o arquivo em uma grande string
text = fp.readlines() #retorna uma lista
xml = "".join(text)

queryRe = re.compile('^orf.{5}')
queryNC = re.compile('NC_.{8}')
queryEv = re.compile('NC_.*')
queryPc = re.compile('Identities = \d*/\d* \(\d*%')
queryEvN = re.compile('\S*$')


sequencias = re.split('Query= ',xml)


for seq in sequencias:
  matchQueyrRe = queryRe.search(seq)
  if matchQueyrRe:
    start = matchQueyrRe.start()
    end = matchQueyrRe.end()
    reFound = seq[start:end]
    result = reFound + ' -'
  #-----
  matchQueryNC = queryNC.search(seq)
  if matchQueryNC:
    start = matchQueryNC.start()
    end = matchQueryNC.end()
    ncFound = seq[start:end]
    result = re.sub(' -',' ' + ncFound,result)
  #-----
  matchQueryEv = queryEv.search(seq)
  if matchQueryEv:
    start = matchQueryEv.start()
    end = matchQueryEv.end()
    evFound = seq[start:end]
    evFound = evFound.rstrip()
    result = result + ' ' + evFound[-3:]
    matchQueryEvN = queryEvN.search(evFound)
    if matchQueryEvN:
      start = matchQueryEv.start()
      end = matchQueryEv.end()
      evNFound = evFound[start:end]
  #-----
  matchQueryPc = queryPc.search(seq)
  if matchQueryPc:
          start = matchQueryPc.start()
          end = matchQueryPc.end()
          pcFound = seq[start:end]
          pcFound = re.sub(r'Identities = \d*/\d* \(','',pcFound)
          result = result + ' ' + pcFound
  print(result)
```
