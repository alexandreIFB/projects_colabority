#Motif D

## Leitura do arquivo e montagem da matriz

```python
filename = '/content/motif_d.fasta'

dictSeqs = {}

with open(filename) as fp:
    for linha in fp:
      if(linha[0] == '>'):
        header = linha.strip()
        dictSeqs[header] = ''
      else:
        linha.rstrip()
        dictSeqs[header] = dictSeqs[header] + linha.replace('\n','')

#for i in dictSeqs.values():
  #print(i)

letras = ['A','U','G','C']

#print(len(letras))

matriz = [['x',1,2,3,4,5,6,'Freq Overall']]
linhaCont = 0
while(linhaCont < len(letras)):
  linha = []
  linha.append(letras[linhaCont])
  for x in range(0,7):
    linha.append('0')
  matriz.append(linha)
  linhaCont += 1


print(len(matriz))
```

```python
import numpy as np
import pandas as pd


matriz = np.array(matriz)
# panda_df = pd.DataFrame(data = matriz[1:, 1:],
#                         index = matriz[1:, 0],
#                         columns = matriz[0, 1:])

# # Mostra a matriz pontuação
# display(panda_df)
```

## Frequencia na matriz

```python
for x in range(0,6):
  tabela = {
  "A":0, "U":0, "G":0, "C":0,
  }
  for i in dictSeqs.values():
    tabela[i[x]] += 1
  freq = [tabela['A'] / 72,tabela['U'] / 72,tabela['G'] / 72,tabela['C'] / 72]
  matriz[0+1,x+1] = freq[0]
  matriz[1+1,x+1] = freq[1]
  matriz[2+1,x+1] = freq[2]
  matriz[3+1,x+1] = freq[3]
```

```python
panda_df = pd.DataFrame(data = matriz[1:, 1:],
                        index = matriz[1:, 0],
                        columns = matriz[0, 1:])

panda_df = panda_df.astype(float)
panda_df.dtypes

#display(panda_df)

for i in range(0,4):
  panda_df.iat[i,6] = panda_df.iloc[i,0:-1].mean()

copiaMatriz = panda_df.copy()

copiaSubistituida = panda_df.copy()

display(panda_df)


```

## Matriz normalizada

```python
for i in range(0,4):
  for x in range(0,6):
    panda_df.iat[i,x] = panda_df.iat[i,x] / panda_df.iat[i,6]

display(panda_df)
```

## Substituindo por min - media

```python
for i in range(0,4):
  for x in range(0,6):
    if(copiaMatriz.iat[i,x] == 0):
      copiaSubistituida.iat[i,x] = min(numero for numero in copiaMatriz.iloc[0:,x] if numero != 0) - copiaMatriz.iloc[0:,x].mean()

display(copiaSubistituida)
```

## Log de 02 sem substituir

```python
np.log2(panda_df)
```

## Log 02 com valors 0 substituidos

```python
for i in range(0,4):
  for x in range(0,6):
    copiaSubistituida.iat[i,x] = copiaSubistituida.iat[i,x] / copiaSubistituida.iat[i,6]

np.log2(copiaSubistituida)
```
